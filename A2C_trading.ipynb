## IMPORT LIBRARIES

```python
import gym
import gym_anytrading
from stable_baselines.common.vec_env import DummyVecEnv
from stable_baselines import A2C
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import yfinance as yf
```

## IMPORT DATASET

```python
df = yf.download('AAPL', start = '2019-01-01', end = '2020-01-01')
df.head()
```

# **1. USING ONLY OHLC DATA**

### 1.1. CREATE & RUN THE ENVIRONMENT

```python
env = gym.make('stocks-v0', df=df, frame_bound=(5,100), window_size=5)
```

```python
state = env.reset()
while True: 
    action = env.action_space.sample()
    n_state, reward, done, info = env.step(action)
    if done: 
        print("info", info)
        break
        
plt.figure(figsize=(15,6))
plt.cla()
env.render_all()
plt.show()
```

### 1.2. TRAINING

```python
env_maker = lambda: gym.make('stocks-v0', df=df, frame_bound=(5,100), window_size=5)
env = DummyVecEnv([env_maker])
```

```python
model = A2C('MlpLstmPolicy', env, verbose=1) 
model.learn(total_timesteps=10000)
```

### 1.3.TESTING

```python
env = gym.make('stocks-v0', df=df, frame_bound=(90,110), window_size=5)
obs = env.reset()
while True: 
    obs = obs[np.newaxis, ...]
    action, _states = model.predict(obs)
    obs, rewards, done, info = env.step(action)
    if done:
        print("info", info)
        break
```

```python
plt.figure(figsize=(15,6))
plt.cla()
env.render_all()
plt.show()
```

### 2.2. CREATE INDICATOR COLUMNS

```python
from gym_anytrading.envs import StocksEnv
from finta import TA
```

```python
df1 = yf.download('AAPL', start = '2019-01-01', end = '2020-01-01')
df1['SMA'] = TA.SMA(df, 12)
df1['RSI'] = TA.RSI(df)
df1['OBV'] = TA.OBV(df)
df1.fillna(0, inplace=True)
```

```python
df1.head()
```

### 2.2. CREATE ENVIRONMENT

```python
def add_signals(env):
    start = env.frame_bound[0] - env.window_size
    end = env.frame_bound[1]
    prices = env.df.loc[:, 'Low'].to_numpy()[start:end]
    signal_features = env.df.loc[:, ['Low', 'Volume','SMA', 'RSI', 'OBV']].to_numpy()[start:end]
    return prices, signal_features
```

```python
class MyCustomEnv(StocksEnv):
    _process_data = add_signals
    
env2 = MyCustomEnv(df=df1, window_size=12, frame_bound=(12,50))
```

### 2.2. TRAINING

```python
env_maker = lambda: env2
env = DummyVecEnv([env_maker])
```

```python
model = A2C('MlpLstmPolicy', env, verbose=1) 
model.learn(total_timesteps=1000000)
```

### 2.3. TESTING

```python
env = MyCustomEnv(df=df1, window_size=12, frame_bound=(80,250))
obs = env.reset()
while True: 
    obs = obs[np.newaxis, ...]
    action, _states = model.predict(obs)
    obs, rewards, done, info = env.step(action)
    if done:
        print("info", info)
        break
```

```python
plt.figure(figsize=(15,6))
plt.cla()
env.render_all()
plt.show()
```
